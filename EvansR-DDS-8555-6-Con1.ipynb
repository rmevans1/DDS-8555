```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
plt.figure(figsize=(8, 8))
plt.title('Recursive Binary Splitting for Airline Delay Prediction')
plt.xlabel('Departure Time (hours)')
plt.ylabel('Distance (miles)')

# Draw partition lines (cutpoints)
t1 = 10  # Cut at 10 AM (vertical)
t2 = 15  # Cut at 3 PM (vertical)
t3 = 500 # Cut at 500 miles (horizontal)
t4 = 800 # Cut at 800 miles (horizontal)

plt.axvline(x=t1, color='blue', linestyle='--', label='t1: Departure Time = 10 AM')
plt.axvline(x=t2, color='green', linestyle='--', label='t2: Departure Time = 3 PM')
plt.axhline(y=t3, color='red', linestyle='--', label='t3: Distance = 500 miles')
plt.axhline(y=t4, color='purple', linestyle='--', label='t4: Distance = 800 miles')

# Annotate regions (R1 to R6)
regions = {
    'R1': (5, 650),  # Before t1, between t3 and t4
    'R2': (5, 300),  # Before t1, below t3
    'R3': (12, 650), # Between t1 and t2, between t3 and t4
    'R4': (12, 300), # Between t1 and t2, below t3
    'R5': (17, 650), # After t2, between t3 and t4
    'R6': (17, 300)  # After t2, below t3
}

for region, coord in regions.items():
    plt.text(coord[0], coord[1], region, fontsize=12, fontweight='bold', color='black')

# Set axis limits
plt.xlim(0, 24)
plt.ylim(0, 1000)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
