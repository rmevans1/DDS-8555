## Load Data

```python
Auto = load_data('Auto')
Auto.columns
```

```python
def plot_fit(the_df, x_data, basis, title, x_axis):
  X = basis.transform(Auto)
  Xnew = basis.transform(the_df)
  M = sm.OLS(y, X).fit()
  preds = M.get_prediction(Xnew)
  bands = preds.conf_int(alpha=0.05)
  fig, ax = subplots(figsize=(8,8))
  ax.scatter(x_data, y, facecolor='gray', alpha=0.5)

  for val, ls in zip([preds.predicted_mean,
                      bands[:,0],
                      bands[:,1]],
                     ['b', 'r--', 'r--']):
    ax.plot(the_df.values, val, ls, linewidth=3)
    ax.set_title(title, fontsize=20)
    ax.set_xlabel(x_axis, fontsize=20)
    ax.set_ylabel('MPG', fontsize=20)
    return ax
```

## Data Exploration

```python
# Select numerical columns (excluding 'name' and others if necessary)
numerical_cols = Auto.select_dtypes(include=['number']).columns
numerical_cols = numerical_cols.drop('mpg')

# Calculate the number of rows and columns for the subplots
num_cols = 3
num_rows = (len(numerical_cols) + num_cols - 1) // num_cols  # Ceiling division

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.ravel()  # Flatten the axes array for easier iteration

# Iterate through numerical columns and create scatter plots
for i, col in enumerate(numerical_cols):
    axes[i].scatter(Auto[col], Auto['mpg'])
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('mpg')
    axes[i].set_title(f'{col} vs mpg')

# Turn off any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

## Model Creation- Horsepower

```python
X = Auto['horsepower']
y = Auto['mpg']
```

```python
models = [MS([poly('horsepower', degree=d)]) for d in range(1,8)]
Xs = [model.fit_transform(Auto) for model in models]
anova_lm(*[sm.OLS(y, X_).fit() for X_ in Xs])
```

```python
poly_hp = MS([poly('horsepower', degree=2)]).fit(Auto)
M = sm.OLS(y, poly_hp.transform(Auto)).fit()
summarize(M)
```

```python
hp_grid = np.linspace(X.min(),
                       X.max(),
                       100)
hp_df = pd.DataFrame({'horsepower': hp_grid})
```

```python
plot_fit(hp_df, X, poly_hp, 'Degree-2 Polynomial', 'Horsepower')
```

```python
print(M.summary())
```
