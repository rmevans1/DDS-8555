```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients
beta_0 = 1
beta_1 = 1
beta_2 = -2

# Define the basis functions
def b1(x):
    return x

def b2(x):
    return (x - 1) ** 2 * (x >= 1)

# Define the regression model
def regression_model(x):
    return beta_0 + beta_1 * b1(x) + beta_2 * b2(x)

# Generate X values between -2 and 2
x_values = np.linspace(-2, 2, 500)
y_values = regression_model(x_values)

# Plot the estimated curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Estimated Curve", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.7)  # x-axis
plt.axvline(0, color="black", linestyle="--", linewidth=0.7)  # y-axis

# Highlight intercepts and other key points
plt.scatter(0, regression_model(0), color="red", label="Intercept")
plt.scatter(1, regression_model(1), color="green", label="Transition Point")

# Add annotations for clarity
plt.text(0, regression_model(0) + 0.2, f"(0, {regression_model(0):.1f})", color="red")
plt.text(1, regression_model(1) + 0.2, f"(1, {regression_model(1):.1f})", color="green")

# Add labels, title, and legend
plt.title("Estimated Regression Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
